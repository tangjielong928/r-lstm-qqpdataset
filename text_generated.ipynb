```python
import collections
import sys
import numpy as np
import nltk
from nltk.corpus import stopwords
```

```python
with open('src_test_12000.txt', 'r', encoding='UTF-8') as f:
    f_lines = f.readlines()
    src = [w.strip() for w in f_lines]

with open('tgt_test_12000.txt', 'r', encoding='UTF-8') as f:
    f_lines = f.readlines()
    tgt = [w.strip() for w in f_lines]

with open('jl_qqp_12000_gen.txt', 'r', encoding='UTF-8') as f:
    f_lines = f.readlines()
    pred = [w.strip() for w in f_lines]
```

```python
comp =[]
for s,t in zip(src,tgt):
    comp.append(s+'\t'+t)
```

```python
final = []
for c,pre in zip(comp,pred):
    final.append(c+'\t'+pre+'\n')
```
